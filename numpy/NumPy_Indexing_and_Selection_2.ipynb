## Bracket Indexing and Selection
The simplest way to pick one or some elements of an array looks very similar to python lists:

```python
#Creating sample array
arr=np.arange(10,100,5)
arr
```

```python
len(arr)
```

```python
arr[-1]
```

```python
#Get a value at an index
arr[9]
```

```python
#Get values in a range
arr[1:11:2]
```

## Indexing a 2D array (matrices)

The general format is **arr_2d[row][col]** or **arr_2d[row,col]**. It is recommended to use the comma notation for clarity.

```python
import numpy as np
```

```python
arr_2d = np.array(([1,2,3],[12,15,18],[64,96,128]))

#Show
arr_2d
```

```python
# syntex for 2d array indexing 
#[row,col],[row][col]
```

```python
arr_2d[2][2]
```

```python
arr_2d[:,1]
```

```python
l1=[1,2,3,4,5]
```

```python
l1[:3]
```

```python
arr_2d[:2,:2]
```

```python
arr3=np.arange(27).reshape(3,3,3)
```

```python
arr3
```

```python
arr_2d[1:,:2]
```

```python
arr_2d.shape
```

```python
arr4=np.arange(15).reshape(3,5)
```

```python
arr4
```

```python
arr4[1:,1:4]
```

```python
arr_2d[1]  # Indexing 2nd row
```

```python
arr_2d[:,2] # Indexing 3rd column
```

```python
arr_2d[1:,1:]
```

```python
arr_2d[1,1]
```

```python
arr_2d[0:2][0:2]
```

```python
# Indexing column
arr_2d[:,2]
```

```python
arr5=np.arange(20).reshape(5,4)
```
